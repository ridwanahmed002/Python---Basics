# For Loop  

```python
from IPython.display import Image
Image(filename=r"D:\blog pics\for-loop-python.png")
```

```python
# Normally

name = "Ridwan"
for i in name:
    print(i, 'i')
```

```python
# using range

name = "Rid wan"
for i in range(len(name)):
    print(name[i], 'i')
```

```python
# adding numbers in a list

L = [1,2,3,4,5]
sum = 0
for i in L:
    sum += i
    print(i, sum)
print('sum is = ' , sum)
```

# while loop    

```python
from IPython.display import Image
Image(filename=r"D:\blog pics\while-loop.png")
```

```python
name = "Rid wan"
i = 0
while i < len(name):
    print(name[i], 'i')
    i += 1
```

# Nested Loop

```python
# nested loop

for i in range(1,5):
    for j in range(1,5):
        print(i,j)
    print()
```

```python
# easy nested loop example with symbol

for i in range(1,5):
    for j in range(1,5):
        print(i,'*',j,'=',i*j)
    print()
```
